## Case of first-order plant

```python
import numpy as np
import control as ctrl
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['font.family'] = ['fantasy']
plt.rcParams['font.fantasy'] = ['Comic Sans MS']
```

Let's consider the operational operator `s`

```python
s = ctrl.tf('s')
```

We suppose, hereafter to investigate the performance of first-order plant, subject to external disturbance. Its transfer function is given by $$\mathcal{H}(s) \;=\; \dfrac{k}{1+\tau\, s}$$

```python
H = 1/(s+1)
```

The controller we will be using, has the following transfer function
$$\mathcal{G}(s) \;=\; k_p+\dfrac{k_i}{s}$$

```python
R = 1 + 1/s
```

Considering a unit feedback yields the following closed 

```python
FTBF = ctrl.feedback(ctrl.series(R, H), 1)
FTBF
```

```python
tspan, av = ctrl.step_response(FTBF)
```

```python
plt.plot(tspan, av)
plt.xlabel('t (sec)')
plt.grid()
```
